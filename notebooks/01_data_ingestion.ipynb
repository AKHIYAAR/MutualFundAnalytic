# 01 Data Ingestion Pipeline

This notebook tracks and documents the initial raw data ingestion process for the mutual fund schemes, historical NAV files, and investor transactions/demographics.

```python
import os
import pandas as pd
import glob

# Print raw CSV files
raw_files = glob.glob('Data/raw/*.csv')
print(f"Found {len(raw_files)} raw CSV files in Data/raw:")
for f in raw_files:
    print(f"  - {os.path.basename(f)} (Size: {os.path.getsize(f) / 1024:.2f} KB)")
```

### Ingestion Quality Check
We inspect the column structure of the main raw files to ensure referential integrity before database schema design.

```python
for f in ['Data/raw/fund_master.csv', 'Data/raw/nav_history.csv', 'Data/raw/investor_transactions.csv']:
    if os.path.exists(f):
        df = pd.read_csv(f, nrows=5)
        print(f"\nColumns for {os.path.basename(f)}:")
        print(df.columns.tolist())
        print(df.head(2))
```
