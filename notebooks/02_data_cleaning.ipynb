# 02 Data Pre-processing & Cleaning

This notebook documents the preprocessing steps taken to clean historical daily NAV data and investor transactions. 

### Cleaning Steps:
1. **Date Standardisation**: Converting all transaction and NAV dates to standard datetime objects.
2. **Decimal Correction**: Multiplying NAV entries before 30-08-2015 by 100 for `axis_bluechip_119092.csv` (correcting 100x shift).
3. **Zero NAV Interpolation**: Cleaning zero-NAV values for `icici_bluechip_120503.csv` on 07-04-2013 using linear interpolation.
4. **Reindexing and Forward-Filling**: Reindexing to complete daily calendar range to handle weekends/holidays, then forward-filling NAV.
